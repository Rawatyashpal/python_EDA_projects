```python
import pandas as pd
import numpy as np
import random
num_rows = 25000
names = ['ramveer', 'karamveer', 'danveer', 'balveer', 'shaktiman', 'pokemon', 'Doremon', 'Sophia', 'James', 'Thomas']
genders = ['M', 'F']
countries = ['USA', 'Canada', 'UK', 'Australia', None]
positions = ['Manager', 'Developer', 'Analyst', 'Clerk', 'Intern']
garbage_values = ['?', 'N/A', 'Unknown', None, '']
phone_numbers = ['123-456-7890', '234-567-8901', '(345) 678-9012', '4567890123', '567-890-1234', 'N/A', None]

data = {
    'Name': [random.choice(names) for _ in range(num_rows)],
    'Age': [random.choice([25, 30, 35, 28, 40, None, 'Unknown', 'Twenty-five']) for _ in range(num_rows)],
    'Gender': [random.choice(genders) for _ in range(num_rows)],
    'Salary': [random.choice([45000, 50000, 60000, 55000, 70000, None, 'NaN', 120000]) for _ in range(num_rows)],
    'Address': [random.choice(['123 Main St', '456 Elm St', '789 Oak Ave', '222 Pine St', None, 'Unknown']) for _ in range(num_rows)],
    'Country': [random.choice(countries) for _ in range(num_rows)],
    'Position': [random.choice(positions) for _ in range(num_rows)],
    'Garbage': [random.choice(garbage_values) for _ in range(num_rows)],
    'Phone Number': [random.choice(phone_numbers) for _ in range(num_rows)]
}

rows_num=25000
df=pd.DataFrame(data)
df.to_csv('C:\\Users\\Dell\\Desktop\\trick.csv')
```
