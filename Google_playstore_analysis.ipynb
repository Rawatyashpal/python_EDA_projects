```python
df_cppy['Size']=df_cppy['Size'].str.replace('M','000')
df_cppy['Size']=df_cppy['Size'].str.replace('k','')
df_cppy['Size']=df_cppy['Size'].replace('Varies with device',np.nan)
df_cppy['Size']=df_cppy['Size'].astype(float)
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_cppy[cols]=df_cppy[cols].str.replace(item,'')
```

```python
df_cppy['Price'].unique()
```

```python
df_cppy['Installs'].unique()
```

```python
df_cppy['Installs']=df_cppy['Installs'].astype('int')
df_cppy['Price']=df_cppy['Price'].astype('float')
```

```python
df_cppy['Last Updated'].unique()
```

```python
df_cppy['Last Updated']=pd.to_datetime(df_cppy['Last Updated'])

df_cppy['Day']=df_cppy['Last Updated'].dt.day

df_cppy['Month']=df_cppy['Last Updated'].dt.month

df_cppy['Year']=df_cppy['Last Updated'].dt.year
```

```python
df_cppy=df_cppy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_cppy.shape
```

```python
numeric_features=[feature for feature in df_cppy.columns if df_cppy[feature].dtype !=0]
```

```python
numeric_features=[feature for feature in df_cppy.columns if df_cppy[feature].dtype!='O']
categorical_feature=[feature for feature in df_cppy.columns if df_cppy[feature].dtype==0]
print('We have {} numerical feature :{}'.format(len(numeric_features),numeric_features))
print('\nWe have {} categorical features :{}'.format(len(categorical_feature),categorical_feature))
```

```python
for col in numeric_features:
    print(df[col].value_counts(normalize=True)*100)
    print('------------------------------')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,15))
plt.suptitle('niveriate analysis of numerical features',fontsize=20,fontweight='bold',alpha=0.8,y=1)

for i in range(0,len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_cppy[numeric_features[i]],shade=True,color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,15))
plt.suptitle('univeriant analysis of categorical features',fontsize=20,fontweight='bold',alpha=0.8,y=1)
category=['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
